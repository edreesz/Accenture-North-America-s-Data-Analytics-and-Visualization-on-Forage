```python
import pandas as pd
```

```python
contents = pd.read_csv('Content.csv')
reaction = pd.read_csv('Reactions.csv')
reaction_type = pd.read_csv('ReactionTypes.csv')
```

```python
column_drop = ['Unnamed: 0','User ID','URL']
contents.drop(columns=column_drop, inplace=True, errors='ignore')
reaction.drop(columns=column_drop, inplace=True, errors='ignore')
reaction_type.drop(columns=column_drop, inplace=True, errors='ignore')
```

```python
contents = contents.dropna()
reaction = reaction.dropna()
reaction_type = reaction_type.dropna()
```

```python
contents.info(),reaction.info(),reaction_type.info()
```

```python
reaction.rename(columns={'Type':'Reaction Type'}, inplace=True)
reaction_type.rename(columns={'Type':'Reaction Type'}, inplace=True)
contents.rename(columns={'Type':'Content Type'}, inplace=True)
```

```python
reactions_merged = pd.merge(reaction, reaction_type, on='Reaction Type',how='inner')
```

```python
content_reactions = pd.merge(contents, reactions_merged, on='Content ID',how='inner')
```

```python
content_reactions.to_csv('content_reactions_cleaned.csv')
```

```python
category_counts = content_reactions['Category'].value_counts().nlargest(5)
plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Top 5 Content Categories')
plt.xlabel('Category')
plt.ylabel('Number of Posts')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(category_counts.values, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Top 5 Content Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python

top_categories = content_reactions['Category'].value_counts().nlargest(5).index.tolist()


filtered_df = content_reactions[content_reactions['Category'].isin(top_categories)]


heatmap_data = filtered_df.pivot_table(index='Category', columns='Reaction Type', aggfunc='size', fill_value=0)



plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap of Top 5 Categories vs. Reaction Type')
plt.xlabel('Reaction Type')
plt.ylabel('Category')
plt.show()
```
